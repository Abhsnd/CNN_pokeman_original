划分数据集

```python
from sklearn.model_selection import train_test_split

def train_test_val_split(df, ratio_train, ratio_test, ratio_val):
    """
    划分 训练集 测试集 验证集
    df: 要划分数据集
    ratio_train: 训练集比例
    ratio_test: 测试集比例
    ratio_val: 验证集比例
    """
    train, middle = train_test_split(df, test_size=1-ratio_train)
    ratio = ratio_val/(1-ratio_train)
    test,validation = train_test_split(middle, test_size=ratio)
    return train,test,validation
```

```python
# 划分比例 7：1：2 训练集 验证集 测试集
train_bulbasaur, test_bulbasaur, validation_bulbasaur = train_test_val_split(bulbasaur_img_src, ratio_train=0.7, ratio_test=0.2, ratio_val=0.1)
train_charmander, test_charmander, validation_charmander = train_test_val_split(charmander_img_src, ratio_train=0.7, ratio_test=0.2, ratio_val=0.1)
train_mewtwo, test_mewtwo, validation_mewtwo = train_test_val_split(mewtwo_img_src, ratio_train=0.7, ratio_test=0.2, ratio_val=0.1)
train_pikachu, test_pikachu, validation_pikachu = train_test_val_split(pikachu_img_src, ratio_train=0.7, ratio_test=0.2, ratio_val=0.1)
train_squirtle, test_squirtle, validation_squirtle = train_test_val_split(squirtle_img_src, ratio_train=0.7, ratio_test=0.2, ratio_val=0.1)
```

卷积神经网络

```python
from tensorflow.keras import layers
from tensorflow.keras import models
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential()
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))

model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(5, activation='softmax'))

print("[INFO] training network...")
sgd = SGD(0.01)
```

迁移学习

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
base_model = VGG16(weights="imagenet", include_top=False, input_shape=(150, 150, 3))
```

```python
# 自定义顶层网络
top_model = models.Sequential()
top_model.add(layer=layers.Flatten(input_shape=base_model.output_shape[1:])) # 将预训练网络展平
top_model.add(layer=layers.Dense(256, activation="relu"))   # 全连接层，输入像素256
top_model.add(layer=layers.Dropout(0.5))  # 正则化
top_model.add(layer=layers.Dense(5, activation="softmax"))  # 输出层，多分类
```

```python
model1 = models.Model(inputs=base_model.input, outputs=top_model(base_model.output))
```

```python
from tensorflow.keras import optimizers
from tensorflow.keras import losses
from tensorflow.keras import metrics
```

```python
# 构建网络
model.compile(loss='categorical_crossentropy', optimizer=optimizers.RMSprop(lr=1e-4), metrics=['acc'])
```

```python
# 构建迁移学习网络
model1.compile(loss='categorical_crossentropy', optimizer=optimizers.RMSprop(lr=1e-4), metrics=['acc'])
```

开始训练网络

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1/255)
validation_datagen = ImageDataGenerator(rescale=1/255)

train_generator = train_datagen.flow_from_directory(train_dir, target_size=(150,150),
                                                    batch_size=100,class_mode='categorical') # batch_size 一次训练所选取的样本数
validation_generator = validation_datagen.flow_from_directory(validation_dir, target_size=(150,150), 
                                                              batch_size=40, class_mode='categorical')
```

```python
# 卷积神经网络模型训练 训练次数 20次
history = model.fit(train_generator, epochs=20, validation_data=validation_generator)
```

```python
# 迁移学习模型训练 训练次数 4次
history1 = model1.fit(train_generator, epochs=4, validation_data=validation_generator)
```

```python
def show_train_history(train_history, train, validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train History')
    plt.xlabel('Epoch')
    plt.ylabel(train)
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
```

```python
def plot_images_labels_prediction(images, labels, prediction, idx=0, num=10):
    """
    images为图像
    labels为标签
    prediction为预测
    idx为要开始显示的图像的索引，开始为0
    num为要显示图像的数量，默认是10
    """
    fig = plt.gcf()
    fig.set_size_inches(22,24)
    if num > 25:
        num = 25
    for i in range(0, num):
        ax = plt.subplot(5,5,i+1)
        ax.imshow(images[idx], cmap='binary')
        title = str(i) + '.' + label_dict_reverse[labels[i]]
        if len(prediction) > 0:
            title += '=>' + label_dict_reverse[prediction[i][0]]
        ax.set_title(title, fontsize=10)
        ax.set_xticks([])
        ax.set_yticks([])
        idx += 1
    plt.show()
```
